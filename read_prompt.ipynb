```python
import re

with open("prompts.txt", "r") as f:
    prompts = []
    prompts.append({})
    for line in f.readlines():
        if line.strip() == "":
            prompts.append({})
        else:
            if 'base' not in prompts[-1].keys():
                prompts[-1]['base'] = line.strip()
            elif 'promot' not in prompts[-1].keys():    
                prompts[-1]['promot'] = line.strip()
                prompts[-1]['dir_name'] = re.sub(r'[\W_]+', ' ', line.strip()).replace(" ", "_").lower()
                prompts[-1]['additional_prompts'] = []
                prompts[-1]['file_names'] = []
            else:
                prompts[-1]['additional_prompts'].append(line.strip()[2:])
                prompts[-1]['file_names'].append(re.sub(r'[\W_]+', ' ', line.strip()[2:]).replace(" ", "_").lower())

prompts
```
